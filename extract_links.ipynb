```python
import pandas as pd
file = 'chicago.xlsx'
data = pd.read_excel(file, usecols=['URL'],sheet_name = 0,  
                        index_col = 0)
```

```python
import math
final_links = []
for row in data.iterrows():
    if not isinstance(row[0], float):
        for r in row[0].split(', '):
            if r:
                final_links.append(r)
```

```python
final_links
```

```python
final_domain_names = list()

from urllib.parse import urlparse
for link in final_links:
    domain = urlparse(link).netloc
    final_domain_names.append(domain)
```

```python
from collections import Counter

d = Counter(final_domain_names).most_common()
```

```python
import csv
with open('websites_used_for_chicago.csv','w') as csvfile:
    fieldnames=['website','count']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for key, value in d:
        writer.writerow([key, value])
```

```python
import pandas as pd
file = 'compare4.csv'
data = pd.read_csv(file, usecols=['website'],
                        index_col = 0)
```

```python
import math
final_links = []
for row in data.iterrows():
    final_links.append(row[0])
```

```python
final_links
```

```python
from collections import Counter

d = Counter(final_links).most_common()
```

saving result in compare4_csv

```python
import csv
with open('compare4_result.csv','w') as csvfile:
    fieldnames=['website','refereneced counter']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for key, value in d:
        writer.writerow([key, value])
```
