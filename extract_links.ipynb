```python
import pandas as pd
file = './10_universities/8 IMPERIAL COLLEGE LONDON, UK.xlsx'
data = pd.read_excel(file, usecols=['url', 'LinkedIn_id', 'facebook_id', 'twitter_id', 'Research_gate_id', 'Google_scholar_id'],sheet_name = 0,  
                        index_col = None)
```

```python
import math
final_links = []
for i, row in data.iterrows():
    
    row0 = str(row[0]) + "," + str(row[1]) + "," + str(row[2]) + "," + str(row[3])
    if not isinstance(row0, float):
        for r in row0.split(','):
            if r and not r.startswith('Not') and r != 'nan' and r != " ":
#                 print(type(r), r)
                final_links.append(r)
```

```python
final_links
```

```python
final_domain_names = list()

from urllib.parse import urlparse
for link in final_links:
    domain = urlparse(link).netloc
    final_domain_names.append(domain)
    
print(final_domain_names, len(final_domain_names))
```

```python
from collections import Counter

d = Counter(final_domain_names).most_common()
```

```python
import csv
with open('websites_used_for_chicago.csv','w') as csvfile:
    fieldnames=['website','count']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for key, value in d:
        writer.writerow([key, value])
```
