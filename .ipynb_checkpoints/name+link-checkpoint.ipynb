```python
import pandas as pd
file = 'Chicago/chicago.xlsx'
data = pd.read_excel(file, usecols=['Name','URL'],sheet_name = 0,  
                        index_col = 0)
```

```python
import math
import csv
with open('chicago22.csv','w') as csvfile:
    fieldnames=['Name','URL']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for index, row in data.iterrows():
        row0 = row[0].replace(',', ', ')
        writer.writerow([index, row0])
#         for i, r in enumerate(row0.split(',')):
#             if r:
#                 if i == 0:
#                     writer.writerow([index, r])
#                 else:
#                     writer.writerow(['', r])
```
