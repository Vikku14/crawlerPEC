```python
new_r, new_c = new_final.shape
total = new_r * new_c
```

```python
import openpyxl
import math
from openpyxl.styles import PatternFill
wb = openpyxl.load_workbook(new_final_file)
ws = wb['Sheet1']
change_counter=0

fill_pattern = PatternFill(patternType="solid", fgColor='32CD32')


for i in range(new_r):
    for j in range(new_c):
        if new_final.iloc[i,j] != old_data.iloc[i,j]:
            if isinstance(new_final.iloc[i,j], str):
                ws.cell(row=i+2, column=j+1).fill = fill_pattern
                change_counter += 1
            else:
                if not math.isnan(new_final.iloc[i,j]):
                    ws.cell(row=i+2, column=j+1).fill = fill_pattern
                    change_counter += 1
            
            
wb.save('marked.xlsx')

percent_change = (change_counter*100)/total
```
