```python
import pandas as pd
file = '/home/vikku/thesis/general/final/10 UNIVERSITY OF CHICAGO, USA.csv'
data = pd.read_csv(file, usecols=['url'], 
                        index_col = None)
```

```python
import math
final_links = []
for i, row in data.iterrows():
    row0 =row[0]
    for r in row0.split('\n'):
        if r:
            final_links.append(r)
```

```python
# final_links
```

```python
total_websites_hits, twitter_hits, linkedin_hits, scholar_hits, facebook_hits, research_hits = len(final_links), 0, 0, 0, 0, 0
mis_websites = 0
for link in final_links:
    if 'twitter' in link:
        twitter_hits += 1
    elif 'linkedin' in link:
        linkedin_hits += 1
    elif 'scholar' in link:
        scholar_hits += 1
    elif 'facebook' in link:
        facebook_hits += 1
    elif 'researchgate' in link:
        research_hits += 1
    else:
        mis_websites += 1
print(total_websites_hits, twitter_hits, linkedin_hits, scholar_hits, facebook_hits, research_hits, mis_websites)

print(total_websites_hits ==twitter_hits+ linkedin_hits+ scholar_hits+facebook_hits+ research_hits+ mis_websites)
```

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
california = ['LinkedIn','Research Gate', 'Miscellaneous', 'Facebook', 'Twitter', 'Google Scholar']

explode = (0.0, 0.1, 0.0, 0.1, 0.0, 0.1)
# data = [linkedin_hits, research_hits, mis_websites, facebook_hits, twitter_hits, scholar_hits]
data = [1951, 1483,9272, 410,726, 1180]
colors = ['#0A66C2', '#40BA9B', 'orange', '#E4E6EB','#5DA9DD'  ,'#767676']


def func(pct, allvalues):
    
    absolute = int(round(pct / 100.*np.sum(allvalues), 0))
    return "{0} Hits".format(absolute)
# Creating plot

fig, ax = plt.subplots(figsize =(10, 10))
wedges, texts, autotexts = ax.pie(data, labels = california, colors=colors,explode = explode,autopct=lambda pct: func(pct, data))


# Adding legend
ax.legend(wedges, california ,
          title ="Websites",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
  
plt.setp(autotexts, size = 12, weight ="bold")
# ax.set_title("Total Websites Hits: "+ str(total_websites_hits))
ax.set_title("Total Websites Hits: 15022")

# show plot
plt.show()

```
