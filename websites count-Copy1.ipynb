```python
import pandas as pd
file = '/home/vikku/thesis/general/final/1 CALIFORNIA INSTITUTE OF TECHNOLOGY, USA.csv'
data = pd.read_csv(file, usecols=['url'], 
                        index_col = None)
```

```python
import math
final_links = []
for i, row in data.iterrows():
    row0 =row[0]
    for r in row0.split('\n'):
        if r:
            final_links.append(r)
```

```python
final_links
```

```python
total_websites_hits, twitter_hits, linkedin_hits, scholar_hits, facebook_hits, research_hits = len(final_links), 0, 0, 0, 0, 0
mis_websites = 0
for link in final_links:
    if 'twitter' in link:
        twitter_hits += 1
    elif 'linkedin' in link:
        linkedin_hits += 1
    elif 'scholar' in link:
        scholar_hits += 1
    elif 'facebook' in link:
        facebook_hits += 1
    elif 'researchgate' in link:
        research_hits += 1
    else:
        mis_websites += 1
print(total_websites_hits, twitter_hits, linkedin_hits, scholar_hits, facebook_hits, research_hits, mis_websites)

print(total_websites_hits ==twitter_hits+ linkedin_hits+ scholar_hits+facebook_hits+ research_hits+ mis_websites)
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]


# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)

# Creating color parameters
colors = ( "orange", "cyan", "brown",
		"grey", "indigo", "beige")

# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }

# Creating autocpt arguments
def func(pct, allvalues):
	absolute = int(pct / 100.*np.sum(allvalues))
	return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
								autopct = lambda pct: func(pct, data),
								explode = explode,
								labels = cars,
								shadow = True,
								colors = colors,
								startangle = 90,
								wedgeprops = wp,
								textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, cars,
		title ="Cars",
		loc ="center left",
		bbox_to_anchor =(1, 0, 0.5, 1))

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```

```python
import pandas as pd
file = 'compare4.csv'
data = pd.read_csv(file, usecols=['website'],
                        index_col = 0)
```

```python
import math
final_links = []
for row in data.iterrows():
    final_links.append(row[0])
```

```python
final_links
```

```python
from collections import Counter

d = Counter(final_links).most_common()
```

```python
d
```

saving result in compare4_csv

```python
import csv
with open('compare4_result.csv','w') as csvfile:
    fieldnames=['website','refereneced counter']
    writer=csv.writer(csvfile)
    writer.writerow(fieldnames)
    for key, value in d:
        writer.writerow([key, value])
```
